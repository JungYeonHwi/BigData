```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df
```

```python
# 해당 column 값 가지고 오기
df[['sepal_width', 'sepal_length', 'species']]
```

```python
# .을 통해 해당 column을 가지고 올 때는 한글이나 특수 문자 불가능하기에 []로 감싸기
df.sepal_width
```

```python
# 조건에 해당하는 column을 가지고 오기 (조건은 column 명에 해당하는 것)
df.filter(regex='_')
```

```python
# 해당 문자로 끝나는 조건의 cloumn 가지고 오기 
df.filter(regex='length$')
```

```python
# 해당 문자로 시작하는 조건의 cloumn 가지고 오기 
df.filter(regex='^sepal')
```

```python
# 해당 문자로 시작하고, 끝나는 column 가지고 오기
df.filter(regex='^x[1-5]$')
```

```python
# 첫 번째 파라미터는 가지고 올 범위 (해당 인덱스까지)
# 두 번째 파리미트는 어떤 column 부터 가지고 올지 설정
# 세 번째 파라미터는 끝 column 을 설정
df.loc[2:5, "sepal_width" : "petal_width"]
```

```python
# 첫 번째 파라미터는 가지고 올 범위 (해당 인덱스 전까지)
# 두 번째 파리미트는 몇 번째의 column을 가지고 올지
df.iloc[2:5, [1,2,4]]
```

```python
# 
df.loc[df['sepal_length'] > 5, ['sepal_length', 'sepal_width'] ]
```
