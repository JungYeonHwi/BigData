```python
# Hexbin plot은 데이터가 클 때 각각의 점을 산점도(scatter plot)로 표현할 때의 단점을 보완할 수 있습니다.
# https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html#hexagonal-bin-plot
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl

%matplotlib inline

# 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
```

```python
# girdsize로 조절 가능
df.plot.hexbin(x='a', y='b', gridsize=10)
```

```python
# 첫 번재 숫자와 두 번째 숫자 사이의 값으로 골고루 생성
df['z'] = np.random.uniform(0, 3, 1000)
```

```python
# 기본적으로 각 (x, y) 점 주변의 개수에 대한 히스토그램이 계산
# C 및 reduce_C_function 인수에 값을 전달하여 대체 집계를 지정 가능
# C는 각 (x, y) 점에서 값을 지정하고 reduce_C_function은 빈의 모든 값을 단일 숫자 (예 : 평균, 최대 값, 합계, 표준)로 줄이는 하나의 인수의 함수
# 이 예에서 위치는 열 (column) a 및 b에 의해 주어지며 값은 열 (z)에 의해 주어짐
# 빈은 NumPy의 max 함수로 집계됨
df.plot.hexbin(x='a', y='b', C='z', 
               reduce_C_function=np.median, gridsize=20)
```
