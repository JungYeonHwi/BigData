```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
df.head(3)
```

```python
# by를 통하여 어떤 column group화 할지 결정
# size()를 통하여 group화 명령어
df.groupby(by="origin").size()
```

```python
# value_counts()를 통해 위와 동일한 결과
df["origin"].value_counts()
```

```python
# 그룹화 한 뒤 min()을 통해 최솟값
pd.DataFrame(df.groupby(by="origin").min())
```

```python
# 그룹화 한 뒤 max()를 통해 최댓값
pd.DataFrame(df.groupby(by="origin").max())
```

```python
# 그룹화 한 뒤 mean()을 통해 평균값
pd.DataFrame(df.groupby(['model_year', 'origin'])['cylinders'].mean())
```

```python
df2 = pd.DataFrame(
    [[4, 7, 10],
    [5, 11, 8],
    [6, 9, 12]],
    index=[1, 2, 3],
    columns=['a', 'b', 'c'])
```

```python
df2
```

```python
# shift 괄호 숫자만큼 shift 됨
# 마이너스가 들어있으면 반대로 shift 됨
df2['b'].shift(2)
```

```python
# rank() 명령어 : 순위를 보여줌 (밀집도)
# method : 어떤 값을 기준으로 할지 결정 -> min, max 등으로 가능
df["model_year"].rank(method='first').head()
```

```python
# 누적해서 최댓값
df2.cummax()
```

```python
# 누적해서 최솟값
df2.cummin()
```

```python
# 누적해서 곱해짐
df2.cumprod()
```
