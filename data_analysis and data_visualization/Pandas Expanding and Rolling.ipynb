```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
s = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2015', periods=1000))
s
```

```python
# 누적 합을 plot() 으로 시각화
s = s.cumsum()
s.plot()
```

```python
# window 사이즈만큼 이동을 시켜줌
r = s.rolling(window=30)
r.mean()
```

```python
# 추세선
s.plot(style='k--')
# 평균으로 시각화하며 추세선
r.mean().plot(style='k')
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=pd.date_range('1/1/2015', periods=1000),
                  columns=['A', 'B', 'C', 'D'])
df = df.cumsum()
```

```python
df.rolling(window=60).sum().plot(subplots=True)
```

```python
df.rolling(window=len(df), min_periods=1).mean().plot()
```

```python
# expanding을 이용하면 누적해서 보여줌
df.expanding(min_periods=1).mean().plot()
```
